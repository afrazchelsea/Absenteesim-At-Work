### Data
Our task is to build a regression model which will predict the absentieesm in hours based on the employee attributes and information. We will approach it as a Regression Problem.

### Performance Metric
#### RMSE : 

Root Mean Square Error (RMSE) is the standard deviation of the residuals (prediction errors). Residuals are a measure of how far from the regression line data points are, RMSE is a measure of how spread out these residuals are. In other words, it tells you how concentrated the data is around the line of best fit.

#### Missing value analysis 

```python
NA_data = pd.DataFrame(data.isnull().sum())
NA_data = NA_data.rename(columns={0:"NA_sum"})
NA_data["NA_percent"] = (NA_data["NA_sum"]/len(data))*100
NA_data
```

If any feature has more than 30% of missing data then drop that column else perform missing value analysis

```python
for i in range(0,len(NA_data)):
    if(NA_data["NA_percent"][i]>=30):
        data.drop([i],axis = 1)
        # update numerical and categorical set as well
        if(i in numerical_set):
            numerical_set.remove(i)
        else :
            categorical_set.remove(i)
```

```python
# Median method for numerical features
# Mode for categorical features
import statistics
def impute_median_mode(data_set):
    for i in data_set.columns.values:
        if data_set.loc[:,i].dtypes.name == 'int64'or data_set.loc[:,i].dtypes.name == 'float64':
            data_set[i] = data_set[i].fillna(data_set[i].median())
            
        elif data_set.loc[:,i].dtypes.name == 'category':
            data_set[i] = data_set[i].fillna(statistics.mode(data_set[i]))
            
            
    return data_set
```

## Outlier analysis
We will replace outliers with NaN and then impute NAN values with their median or mode values.
Median for numerical features with outliers.
Mode for categorical features with outliers. This will help us achieve normalized distribution

```python
for i in numerical_set:
    print(i)
    q75,q25 = np.percentile(data[i],[75,25])
    iqr = q75-q25
    min_bar = (q25-(1.5*iqr))
    max_bar = (q75+(1.5*iqr))
    print(min_bar,max_bar)
    data.loc[data[i]<min_bar,i] = np.nan
    data.loc[data[i]>max_bar,i] = np.nan
```

```python
data = impute_median_mode(data) # function call  
```

## Model Selection

As this is a regression problem we will be making use of the regression algorithms such as Linear Regression, Decision Tree Regressor and Random Forest Regressor. We will calculate the RMSE score and judge the algorithms based on that.

```python
# Linear Regression 
from sklearn.linear_model import LinearRegression
# Decision trees
from sklearn.tree import DecisionTreeRegressor
# Random forest
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import math
```

```python
modelDT = DecisionTreeRegressor(max_depth=2).fit(df_train.iloc[:,0:14], df_train.iloc[:,14])
```

```python
predictDT = modelDT.predict(df_test.iloc[:,0:14])
```

```python
# RMSE
math.sqrt(mean_squared_error(df_test.iloc[:,14],predictDT))
```

```python
modelRF = RandomForestRegressor(max_depth=2,n_estimators=10,random_state=0,max_features=10).fit(df_train.iloc[:,0:14],\
                                                                                                df_train.iloc[:,14])
predictRF = modelRF.predict(df_test.iloc[:,0:14])
# RMSE
math.sqrt(mean_squared_error(df_test.iloc[:,14],predictRF))
```

```python
lr = LinearRegression()
modelLR = lr.fit(df_train.iloc[:,0:14],df_train.iloc[:,14])
predictLR = modelLR.predict(df_test.iloc[:,0:14])
# RMSE 
math.sqrt(mean_squared_error(df_test.iloc[:,14],predictLR))
```

So amongst the 3, Random Forest gives us the lowest RMSE value of 0.26 whereas Linear Regression gives the highest RMSE value of 0.37 and Decision Tree gives 0.30
